```python
# version 2.0
from IPython.display import clear_output

clear_output(wait=True)

# Inicio del programa
# Solicitar los datos del producto
tipo_producto = int(input("Ingrese el tipo de producto: \n1-Consumo \n2-Productos Farmacéuticos \n3-Bienes Inmuebles \n4-Canasta Familiar\nObs: Productos de la opcion 4-Canasta Familiar son: Arroz, fideos, yerba mate, aceites comestibles, leche, huevo, carnes no cocinadas, harina y sal yodada\n"))
while tipo_producto not in (1,2,3,4):
    clear_output(wait=True)
    print("Favor ingrese un tipo de producto valido: ")
    tipo_producto = int(input("Ingrese el tipo de producto: \n1-Consumo \n2-Productos Farmacéuticos \n3-Bienes Inmuebles \n4-Canasta Familiar\nObs: Productos de la opcion 4-Canasta Familiar son: Arroz, fideos, yerba mate, aceites comestibles, leche, huevo, carnes no cocinadas, harina y sal yodada\n"))
codigo_producto = input("Ingrese el código del producto: ")
while codigo_producto == "" :
    clear_output(wait=True)
    print("Favor ingrese un codigo de producto valido: ")
    codigo_producto = input("Ingrese el código del producto: ")

nombre_producto = input("Ingrese el nombre del producto: ")
while nombre_producto == "":
    clear_output(wait=True)
    print("Favor ingrese un nombre de producto valido: ")
    nombre_producto = input("Ingrese el nombre del producto: ")

precio_unitario = float(input("Ingrese el precio unitario del producto: "))
while precio_unitario <= 0:
    clear_output(wait=True)
    print("El precio debe de ser mayor a 0")
    precio_unitario = float(input("Ingrese el precio unitario del producto: "))

cantidad = int(input("Ingrese la cantidad del producto: "))
while cantidad <= 0:
    clear_output(wait=True)
    print("Ingrese una cantidad valida")
    cantidad = int(input("Ingrese la cantidad del producto: "))

# Calcular el subtotal
subtotal = precio_unitario * cantidad

# Calcular el IVA
if tipo_producto == 1:
    iva = subtotal * 0.10
else:
    iva = subtotal * 0.05

# Calcular el total
total = subtotal + iva

# Mostrar los resultados al usuario
clear_output(wait=True)
print("Código del producto:", codigo_producto)
print("Nombre del producto:", nombre_producto)
print("Precio unitario:", precio_unitario)
print("Cantidad:", cantidad)
print("Subtotal:", subtotal)
print("IVA:", iva)
print("Total:", total)

# Fin del programa
```

```python
# version 3.0 -- se agrega menu para volver a cargar sin terminar la ejecucion, correccion de precio y valores decimales para cantidad
from IPython.display import clear_output #libreria para limpiar pantalla

while True:
    clear_output(wait=True)
    print("Bienvenido al sistema de Catastro de Productos y Calculo de Impuestos")
    print("1- Ingresar datos del producto")
    print("0- Salir")
    opcion = int(input("Seleccione una opción: "))

    if opcion == 1: # aca se pregunta si la opcion es la carga de productos
        clear_output(wait=True)
        print("Ingrese el tipo de producto:")
        print("1- Consumo")
        print("2- Productos Farmacéuticos")
        print("3- Bienes Inmuebles")
        print("4- Canasta Familiar")
        print("Obs: Productos de la opción 4-Canasta Familiar son: Arroz, fideos, yerba mate, aceites comestibles, leche, huevo, carnes no cocinadas, harina y sal yodada")

        tipo_producto = int(input("Seleccione una opción: "))
        while tipo_producto not in (1, 2, 3, 4): # validacion 2 de tipo_producto - valores validos dentro de los tipos existentes
            clear_output(wait=True)
            print("Favor ingrese un tipo de producto válido.")
            tipo_producto = int(input("Seleccione una opción: "))

        codigo_producto = input("Ingrese el código del producto: ")
        while codigo_producto == "": # validacion 1 de codigo_producto - se valida que no sea vacio
            clear_output(wait=True)
            print("Favor ingrese un código de producto válido.")
            codigo_producto = input("Ingrese el código del producto: ")
        while not codigo_producto.isdigit(): # validacion 2 de codigo_producto - se valida que el codigo sea solo numerico
            clear_output(wait=True)
            print("Favor ingrese un código de producto válido (solo números).")
            codigo_producto = input("Ingrese el código del producto: ")

        nombre_producto = input("Ingrese el nombre del producto: ")
        while nombre_producto == "": # validacion 1 de nombre_producto - se valida que no sea un valor vacio
            clear_output(wait=True)
            print("Favor ingrese un nombre de producto válido.")
            nombre_producto = input("Ingrese el nombre del producto: ")

        precio_unitario = float(input("Ingrese el precio unitario del producto: "))
        while precio_unitario <= 0: # validacion 1 de precio_unitario - se valida que no sea un valor negativo
            clear_output(wait=True)
            print("El precio debe ser mayor a 0.")
            precio_unitario = float(input("Ingrese el precio unitario del producto: "))

        cantidad = float(input("Ingrese la cantidad del producto (Kg/Litros/Unidades): ")) # se agrego el tipo de dato float para que admita otras unidades de medida
        while cantidad <= 0:  # validacion 1 de cantidad - se valida que no sea un valor negativo
            clear_output(wait=True)
            print("Ingrese una cantidad válida.")
            cantidad = float(input("Ingrese la cantidad del producto: "))

        # Calcular el subtotal
        subtotal = precio_unitario * cantidad

        # Calcular el IVA
        if tipo_producto == 1:
            iva = subtotal * 0.10
        else:
            iva = subtotal * 0.05

        # Calcular el total
        total = subtotal + iva

        clear_output(wait=True)
        print("Código del producto:", codigo_producto)
        print("Nombre del producto:", nombre_producto)
        print("Precio unitario:", precio_unitario)
        print("Cantidad:", cantidad)
        print("Subtotal:", subtotal)
        print("IVA:", iva)
        print("Total:", total)

        input("Presione Enter para continuar...")

    elif opcion == 0: # aca se sale del menu y del proceso de carga
        break

    else: # aca limpiamos pantalla por si se carga otra opcion que no sea 1 o 0 del menu principal
        clear_output(wait=True)
        print("Opción inválida. Por favor, seleccione una opción válida.")
        input("Presione Enter para continuar...")
```
